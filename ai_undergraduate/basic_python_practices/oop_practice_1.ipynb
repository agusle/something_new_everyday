1. Crear una clase madre Vehiculo con atributos de instancia color y cantidad de ruedas
2. Crear una clase Coche que sea hija de Vehiculo y agregarle atributos de instancia: velocidad y cilindrada


```python
class Vehiculo:
  """
  Es una clase que representa un vehículo.

  Atributos
  ---------
  color = str
    Color del vehículo. Ej: "Rojo"
  
  ruedas = int
    Cantidad de ruedas del vehículo. Ej: 4
  """
  
  def __init__(self,color,ruedas):
    self.color = color
    self.ruedas = ruedas
  
```

Utilizando esta nueva técnica extiende la clase Vehiculo y realiza la siguiente implementación:
image.png

```python
class Coche(Vehiculo):
  """
  Clase que representa un coche, hija de la clase Vehículo.

  Atributos
  --------- 
  velocidad = int
    Velocidad maxima del coche en kilometros/hora. Ej: 120

  cilindrada = int
    Cilindrada del motor del coche en centimetros cubicos. Ej: 3000

  color = str
    Color del vehículo. Ej: "Rojo"
  
  ruedas = int
    Cantidad de ruedas del vehículo. Ej: 4
  """

  def __init__(self,velocidad:int,cilindrada:int,*args,**kwargs):
    """
    El constructor de la clase Coche.
    Hereda los atributos de la clase vehículo y le agrega 
    velocidad y cilindrada a los mismos.
    """
    
    super().__init__(*args,**kwargs)
    self.velocidad = velocidad
    self.cilindrada = cilindrada

class Bicicleta(Vehiculo):
  """
  Clase que representa una bicicleta, hija de la clase Vehículo.

  Atributos
  --------- 
  tipo = str
    Tipo de bicicleta: Urbana o deportiva. Ej: "Urbana"

  color = str
    Color del vehículo. Ej: "Rojo"
  
  ruedas = int
    Cantidad de ruedas del vehículo. Ej: 2
  """

  def __init__(self,tipo:str,*args,**kwargs):
    """
    El constructor de la clase Bicicleta.
    Hereda los atributos de la clase vehículo y le agrega tipo.
    """

    super().__init__(*args,**kwargs)
    self.tipo = tipo
```

```python
class Camioneta(Coche):
  """
  Clase que representa una Camioneta, hija de la clase Coche y
  nieta de la clase Vehiculo.

  Atributos
  --------- 
  carga = int
    Carga total soportada por la camioneta en Kilogramos. Ej: 20000

  velocidad = int
    Velocidad maxima del coche en kilometros/hora. Ej: 120

  cilindrada = int
    Cilindrada del motor del coche en centimetros cubicos. Ej: 3000

  color = str
    Color del vehículo. Ej: "Rojo"
  
  ruedas = int
    Cantidad de ruedas del vehículo. Ej: 4
  """

  def __init__(self,carga,*args,**kwargs):
    """
    El constructor de la clase Camioneta.
    Hereda los atributos de la clase Coche y Vehículo. 
    Le agrega el atributo carga.
    """  

    super().__init__(*args,**kwargs)
    self.carga = carga

class Motocicleta(Bicicleta):
  """
  Clase que representa una Motocicleta, hija de la clase Bicicleta y
  nieta de la clase Vehiculo.

  Atributos
  --------- 
  velocidad = int
    Velocidad maxima del coche en kilometros/hora. Ej: 160

  cilindrada = int
    Cilindrada del motor del coche en centimetros cubicos. Ej: 250

  tipo = str
    Tipo de bicicleta: Urbana o deportiva. Ej: "deportiva"

  color = str
    Color del vehículo. Ej: "amarillo"
  
  ruedas = int
    Cantidad de ruedas del vehículo. Ej: 2
  """
  
  def __init__(self,velocidad,cilindrada,*args,**kwargs):
    super().__init__(*args,**kwargs)
    self.velocidad = velocidad
    self.cilindrada = cilindrada
```

```python
vehiculos = [Camioneta(20000,90,3000,"rojo",4),Motocicleta(160,250,"deportiva","amarillo",2)] 
```


2. Realiza una función llamada catalogar() que reciba la lista de vehiculos y los recorra mostrando el nombre de su clase y sus atributos.

```python
def catalogar(listado:list):
  """
  Recibe un listado de objetos y los recorre mostrando el nombre de su clase,
  sus atributos y sus valores.

  Parameters
  ----------
  listado = list
    Listado de objetos. Ej: [Camioneta(20000,90,3000,"rojo",4)]

  Returns
  -------
  Imprime en pantalla el resultado 
  """
  
  print("Vehículos del listado:")
  for vehiculo in vehiculos:
    print(f"\n{type(vehiculo).__name__}")
    for atributo , valor in vehiculo.__dict__.items():
      print(f"- {atributo}: {valor}")
```

```python
def catalogar(listado:list,ruedas:int=None):
  """
  Recibe un listado de objetos y los recorre mostrando el nombre de su clase,
  sus atributos y sus valores. Si se ingresa valor en el argumento ruedas,
  se van a imprimir solo aquellos objetos cuyo atributo coincida con el número
  ingresado, caso contrario, devuelve el listado normal sin filtros.

  Parameters
  ----------
  listado = list
    Listado de objetos. Ej: [Camioneta(20000,90,3000,"rojo",4)]

  Returns
  -------
  Imprime en pantalla el resultado 
  """

  count = 0
  print("Vehículos del listado:")
  if ruedas != None:
    for vehiculo in vehiculos:
      if vehiculo.__dict__["ruedas"] == ruedas:
        count +=1
        print(f"\n{type(vehiculo).__name__}")
        for atributo , valor in vehiculo.__dict__.items():
          print(f"- {atributo}: {valor}")
    if count > 0:
      print(f"\nSe han encontrado {count} vehículos con {ruedas} ruedas")
    else:
      print(f"\nNo se han encontrado vehículos con {ruedas} ruedas")
  else:
    for vehiculo in vehiculos:
      print(f"\n{type(vehiculo).__name__}")
      for atributo , valor in vehiculo.__dict__.items():
        print(f"- {atributo}: {valor}")
```
